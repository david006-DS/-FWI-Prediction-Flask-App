```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'), 'not fire','fire')
```

```python
# # Recreate fire_data from the original 'df' to retain 'month' column
# temp_df_for_plot = df.copy()
# # Ensure 'Classes' column is in numerical format (0 or 1) as done for dff
# temp_df_for_plot['Classes'] = np.where(temp_df_for_plot['Classes'].str.contains('not fire'), 0, 1)

fire_data = df.loc[df['Region'] == 1] # Filter for fire incidents

plt.subplots(figsize=(15,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=fire_data) # 'month' is now available in fire_data
plt.ylabel("Number of Fires", weight='bold')
plt.xlabel("Month", weight='bold')
plt.title("Fire Analysis of Sidi-Bel Regions", weight='bold')
plt.show()
```

```python
fire_data = df.loc[df['Region'] == 0]

plt.subplots(figsize=(15,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=fire_data)
plt.ylabel("Number of Fires", weight='bold')
plt.xlabel("Month", weight='bold')
plt.title("Fire Analysis of Brjaia Regions", weight='bold')
plt.show()
```

From the above observation, it's observed that August & September had the largest number of forest fires outbreak for both regions. But most of the fire occurred in August rather than in September
